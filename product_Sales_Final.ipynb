<p style="font-size:17px;">To ensure text data was presented in a consistent and readable manner, columns such as Region, Product, StoreLocation, and CustomerName were formatted

```python
dataset.dtypes
```

```python
import re

# Select only string columns
string_cols = dataset.select_dtypes(include=['object']).columns

# Loop through each string column
for col in string_cols:
    
    # Convert column to string to avoid errors
    dataset[col] = dataset[col].astype(str)
    
    # 1. Remove non-printable characters
    dataset[col] = dataset[col].str.replace(r'[\x00-\x1F\x7F]', '', regex=True)
    
    # 2. Trim leading/trailing spaces
    dataset[col] = dataset[col].str.strip()
    
    # 3. Replace multiple spaces with single space
    dataset[col] = dataset[col].str.replace(r'\s+', ' ', regex=True)
```

<p style="font-size:17px;">This line of code removes non-printable characters,removes leading/trailing spaces,replaces multiple spaces with a single 
space.The dataset looked clean, but I still applied these steps to fix any hidden issues if present </p>
